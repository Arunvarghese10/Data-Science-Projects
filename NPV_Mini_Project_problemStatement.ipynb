#### F. If you order grilled chicken rather than crispy chicken, what is the change in calorie intake?
        Step1:  Display the suitable graph for the number of calories in Crispy Chicken category

        Step2: Find the average calories for Crispy Chicken Category

        Step3: Display the suitable graph for the number of calories in Grilled Chicken category

        Step4: Find the average calories for Grilled Chicken Category

        Step5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

        Step6: Write your observations.

##### Step1: Display the suitable graph for the number of calories in Crispy Chicken category

```python
lst1=[]
for i in range(0,len(food['Item_in_Menu'])):
    if 'Crispy Chicken' in food['Item_in_Menu'][i]:
        lst1.append(i)
#print(lst1)
food.iloc[lst1].plot(kind='bar',x='Item_in_Menu',y='Calories')
plt.show()
#len(B['Item_in_Menu'])
```

##### Step2: Find the average calories for Crispy Chicken Category

```python
z=round(food.iloc[lst1].iloc[:,-1].mean(),2)
print("The average calories for crispy chicken category is: ", z)
```

##### Step3: Display the suitable graph for the number of calories in Grilled Chicken category

```python
list2=[]
for i in range(0,len(food["Item_in_Menu"])):
    if "Grilled Chicken" in food["Item_in_Menu"][i]:
        list2.append(i)
        
food.iloc[list2].plot(kind="bar", x="Item_in_Menu", y="Calories")
plt.show()
```

##### Step4:Find the average calories for Grilled Chicken Category

```python
y=round(food.iloc[list2].iloc[:,-1].mean(),2)
print('average calories for Grilled Chicken Category',y)
```

##### Step 5: Get the final visual comparison of Crispy vs Grilled on the basis of calories

```python
Values=[z,y]
labels=['Crispy Chicken','Grilled Chicken']
plt.bar(x=labels,height=Values)
plt.title('Average Calories')
plt.show()
```

##### Step6: Write your observations

This bar graph depicts the data between the two categories of food that is grilled chicken and crispy chicken
with their calories. On the y axis the calories count is displayed for both of the food categories. According to
the data shown in the bar graph average calories for Grilled Chicken Category is 468.33 and average calories for 
Crispy Chicken Category is 474.29. 


Crispy chicken have a slightly more calory intake as compared to grilled chicken.

Thus, Grilled chicken can be considered as comparitively healthier chicken patty w.r.t Crispy chicken.
And the same can be used as a advertising technique to increase the frequency of sales of the same.
